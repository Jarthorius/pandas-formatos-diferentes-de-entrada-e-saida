# Importando direto do Google Planilhas

Para conseguirtrabalhar com a planilha online do google, é precisa:

- Obter o link de compartilhamento da planilha
    - Acessar a planilha e clicar no botão compartilhar
    - Mudar o acesso para qualquer pessoa que tenha o link
    - Copiar o link gerado

- Encontrar dentro do link gerado qual o __ID da Planilha__
    - O ID está depois de "/d/" e antes de "/edit"

        /d/_ID-da-Planilha_/edit

- Atribuir o link a uma variável
- Concatenar a primeira parte do link original com o ID com um parâmetro após:

        /gviz/tq?tqx=out:csv&sheet

    - /gviz/tq? --> API de integração do Google
    - tqx=out:csv --> Tratar a planilha como arquivo csv
    - &sheet --> Necessário para visualizar o cabeçalho da planilhas

```python
sheet_id = '1NfT-Z59-9d-SRlBpqjSENpMwpLMogLbpgRmtPH40LfQ'
```

```python
url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet'
pd.read_csv(url)
```

# Manipulando arquivos json
https://pandas.pydata.org/docs/reference/api/pandas.read_json.html

```python
dados_pacientes = pd.read_json('./dados/pacientes.json')
dados_pacientes.head()
```

```python
dados_pacientes_2 = pd.read_json('./dados/pacientes_2.json')
dados_pacientes_2
```

Para o caso de arquivo json com informações aninhadas, como é o caso do "dados_pacientes_2", é possível desaninhar utilizando função do pandas:

    json_normalize

É possível ver como o dado está organizado em:

    https://jsoncrack.com/

```python
dados_pacientes_2_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
dados_pacientes_2_normalizado
```

Dessa forma, a coluna pacientes foi desaninhada, mas as colunas "Pesquisa" e "Ano" não aparecem. Para que elas apareçam, é necessário:

Importante: Na aula nós realizamos a normalização em um arquivo com o formato JSON. No entanto, a função json_normalize() aceita apenas um dicionário ou uma lista de dicionários.

Para transformar o arquivo json em um dicionário Python é preciso:

- Importar módulo json
- abrir o arquivo pelo path correto e colocá-lo para leitura
- usar o método json.loads()

```python
# transformando arquivo json em dicionário
import json
with open('./dados/pacientes_2.json', 'r') as f:
    dados = json.loads(f.read())

type(dados)
```

```python
dados
```

```python
dados_pacientes_2_normalizado = pd.json_normalize(data = dados, record_path=['Pacientes'], meta=['Pesquisa', 'Ano'])
dados_pacientes_2_normalizado
```

# Exportando DataFrame para arquivo json

```python
dados_pacientes_2_normalizado.to_json('historico_pacientes_normalizado.json')
```

## Criando banco de dados

```python
from sqlalchemy import create_engine, inspect, MetaData, Table, text

engine = create_engine('sqlite:///:memory:')
```

Foi criado o banco de dados. Ele será armazenado dentro da memória local, por isso o parâmetro 'sqlite:///:memory:' foi colocado dentro da função _create_engine()_.

## Alimentando banco de dados

- Importando os dados para o python

```python
caminho = './dados/clientes_banco.csv'
dados = pd.read_csv(caminho)
dados.head()
```

- Exportando para o banco de dados

```python
dados.to_sql('clientes', engine, index=False)
```

- Verificando se a tabela foi criada, isto é, se os dados foram postos no banco de dados

```python
inspector = inspect(engine)
print(inspector.get_table_names())
```

## Lendo uma consulta SQL

- Criar a consulta na linguagem SQL

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

- Realizar a leitura da consulta utilizando o Pandas

```python
empregados = pd.read_sql(query, engine)
empregados
```

## Exportando a consulta SQL para uma nova tabela

```python
empregados.to_sql('empregados', con=engine, index=False)
```

## Lendo uma tabela SQL

```python
pd.read_sql_table('empregados', engine)
```

- Selecionando colunas específicas para a tabela SQL

```python
pd.read_sql_table(table_name='empregados', 
                  con=engine,
                  columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

# Desafio

Chegou a hora de testar os conhecimentos desenvolvidos durante a aula.

Você é responsável por criar um banco de dados local de clientes para uma instituição financeira. Temos o arquivo CSV com os dados de clientes.

Sua missão é:

- Criar o banco de dados local com a biblioteca SQLAlchemy.
- Escrever os dados do arquivo CSV neste banco de dados local.
- Realizar três atualizações no banco de dados:
    - Atualizar o registro do cliente de ID 6840104 que teve o rendimento anual alterado para 300000.
    - Excluir o registro do cliente de ID 5008809, pois essa pessoa não possui mais conta na instituição financeira.
    - Criar um novo registro de cliente seguindo as especificações abaixo:
        - ID_Cliente: 6850985
        - Idade: 33
        - Grau_escolaridade: Doutorado
        - Estado_civil: Solteiro
        - Tamanho_familia: 1
        - Categoria_de_renda: Empregado
        - Ocupacao: TI
        - Anos_empregado: 2
        - Rendimento_anual: 290000
        - Tem_carro: 0
        - Moradia: Casa/apartamento próprio
    
url: https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv

```python
# criar banco de dados
engine_desafio = create_engine('sqlite:///:memory:')

# escrever dados do csv no banco de dados
url_desafio = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
dados_desafio = pd.read_csv(url_desafio)
dados_desafio.to_sql('clientes_desafio', engine_desafio, index=False)
```

```python
# verificar se banco foi criado
inspector_desafio = inspect(engine_desafio)
print(inspector_desafio.get_table_names())
```

```python
# realizando alterações
## atualizar
query_desafio = 'UPDATE clientes_desafio SET Rendimento_anual="300000" WHERE ID_Cliente=6840222'
with engine_desafio.connect() as conn:
    conn.execute(text(query_desafio))

## deletar
query_desafio = 'DELETE FROM clientes_desafio WHERE ID_Cliente=5008809'
with engine_desafio.connect() as conn:
    conn.execute(text(query_desafio))

## criar
query_desafio = 'INSERT INTO clientes_desafio (ID_Cliente, Idade, Grau_escolaridade, Estado_civil, ' \
    'Tamanho_familia, Categoria_de_renda, Ocupacao, Anos_empregado, ' \
    'Rendimento_anual, Tem_carro, Moradia) ' \
    'VALUES (6850985, 33, "Doutorado", "Solteiro", 1, "Empregado", "TI", ' \
    '2, 290000, 0, "Casa/apartamento próprio")'
with engine_desafio.connect() as conn:
    conn.execute(text(query_desafio))
```
